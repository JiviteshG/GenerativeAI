## FinBert model for sentiment for Quant Assistant
Using FinBERT, to acts as a "Quant Assistant" that classifies the sentiment of financial text, which can be a signal for automated trading strategies.

```python
from transformers import BertTokenizer, BertForSequenceClassification, pipeline

# Load FinBERT: A model specialized for financial sentiment
model_name = "yiyanghkust/finbert-tone" # ProsusAI/finbert
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name, num_labels=3)

# Create a sentiment analysis pipeline
nlp = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

```

```python
print("Quant Portfolio Bot: Provide a financial headline or sentence for sentiment analysis.")

while True:
    financial_text = input("Financial Text: ")
    if financial_text.lower() in ['exit', 'quit']:
        break

    # Process the text
    results = nlp(financial_text)

    # Example Output: [{'label': 'Positive', 'score': 0.99}]
    sentiment = results[0]['label']
    confidence = results[0]['score']

    print(f"Analysis: The sentiment is {sentiment} (Confidence: {confidence:.2%})")
    print("-" * 30)
```
