```python
# In order to use OpenAI models we need to install it
# It is a separate library
%pip install -qU langchain-openai
```

```python
from google.colab import userdata

OPENAI_KEY = userdata.get('OPENAI_KEY')
```

```python
from langchain_openai import ChatOpenAI

model_openai = ChatOpenAI(model="gpt-3.5-turbo", api_key=OPENAI_KEY)
```

```python
# When you want to call a runnable entity or object you use .invoke
model_openai.invoke("what is a transformer in machine learning")
```

## For Groq model using Lnagchain

```python
# In order to use Groq models with langchain we need to install it
# It is a separate library if you want to use it with langchain
%pip install -qU langchain-groq
```

```python
GROQ_API_KEY = userdata.get('GROQ_API_KEY')
```

```python
from langchain_groq import ChatGroq

model_groq = ChatGroq(model = "llama-3.1-8b-instant", temperature = 0.8, api_key = GROQ_API_KEY)
```

```python
model_groq.invoke("write a poem about a cute dog", )
```
